```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df  = pd.read_csv(r"C:\Users\mobin\DATASIENCE\ML PROJECTS\CLUSTERING\PCA\Universities.csv")
```

```python
df.head()
```

```python
new_df = df.drop('Univ' , axis = 1)
```

```python
new_df.head()
```

```python
new_df.describe().T
```

```python
scaler = StandardScaler()
std_data = scaler.fit_transform(new_df)
```

```python
pca = PCA(n_components=None)
data  = pca.fit_transform(std_data)
pca_com = pd.DataFrame(data, columns = ['PC1','PC2','PC3','PC4','PC5','PC6'])
pca_com
```

```python
pca_com.corr()
```

```python
# Loking for eigen values.
var = pca.explained_variance_ratio_
var
```

```python
#Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
#eigen vectors

pca.components_
```

```python
plt.plot(var1 , color = 'red', marker = 'o')
plt.title('Scree plot')
plt.xlabel('PC')
plt.ylabel('Cumulative Variance Ratio')
plt.grid()
```

```python
final_df = pd.concat([pd.DataFrame(pca_com.iloc[:,0:2], columns = ['PC1','PC2']) , df[['Univ']]] , axis = 1)
```
