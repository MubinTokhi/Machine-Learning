```python
tsne = TSNE(n_components=2, perplexity=30)
data_tsne = tsne.fit_transform(X)
```

```python
tsne_df = pd.DataFrame(data_tsne, columns = ['X1','X2'])
```

```python
tsne_df['diagnosis'] = df['diagnosis']
```

```python
tsne_df.head()
```

```python
sns.scatterplot(data = tsne_df , x = 'X1', y = 'X2' , hue = 'diagnosis', palette= 'tab10')
```

```python
tsne1 = TSNE(n_components=3, perplexity=30)
data_tsne1 = tsne1.fit_transform(X)
tsne_df1 = pd.DataFrame(data_tsne1, columns = ['X1','X2', 'X3'])
tsne_df1['diagnosis'] = df['diagnosis']
```

```python
data_tsne1
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap

#axes instance
fig = plt.figure(figsize = (8,12))
ax = Axes3D(fig, auto_add_to_figure = False)
fig.add_axes(ax)

#get colormap from seaborn

cmap = ListedColormap(sns.color_palette('husl' , 256).as_hex())

x  = tsne_df1['X1']
y = tsne_df1['X2']
z = tsne_df1['X3']

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

col = tsne_df1['diagnosis'].map({'M':'r','B':'g'})
sc = ax.scatter(x,y,z, s = 20, c = col, marker = 'o' , cmap = cmap, alpha = 1.0)
#Legend

plt.legend(*sc.legend_elements(), bbox_to_anchor = (1.05 , 1) , loc = 2)
plt.show()
```
