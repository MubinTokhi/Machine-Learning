```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split,cross_val_score, GridSearchCV, RandomizedSearchCV
```

## Importing the Data and Preprocessing EDA

```python
filename = 'https://raw.githubusercontent.com/slmsshk/pima-indians-diabetes.data.csv/main/pima-indians-diabetes.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pd.read_csv(filename, names=names)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
```

```python
dataframe.head()
```

```python
# Spliting the data into test_ train part


x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=10)
```

## Building SVM model

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

## GRIDSEARCH CV

```python
clf = SVC()
#param_grid = [{'kernel':['linear','rbf','poly'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.01] }]
param_grid = [{'kernel':['linear','rbf'],'gamma':[0.5,0.1],'C':[10,0.1] }]
gsv = RandomizedSearchCV(clf,param_grid,cv=6,)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(kernel='linear',C=0.1)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
