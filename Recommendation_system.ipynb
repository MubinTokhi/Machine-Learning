```python
movie_table = df.pivot(
    
    index = 'userId', 
    columns = 'movie', 
    values= 'rating'
)
```

```python
movie_table.head()
```

```python
movie_table.fillna(0, inplace = True)
```

```python
movie_table.head()
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1 - pairwise_distances(movie_table.values , metric='cosine')
```

```python
user_sim
```

```python
np.fill_diagonal(user_sim, 0)
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index = df['userId'].unique()
user_sim_df.columns = df['userId'].unique()
```

```python
user_sim_df
```

```python
for x in user_sim_df.index:

    sim_user = user_sim_df.idxmax(axis= 1)[x]
    print(x, sim_user)
    user_1 = df[df['userId']==x]
    user_2 = df[df['userId']== sim_user]
    movies_watched_by_user1 = list(set(user_1['movie']))
    movies_watched_by_user2 = list(set(user_2['movie']))



    for movie_name in movies_watched_by_user1:
        if movie_name not in movies_watched_by_user2:
            print('Recommendation for UserID', sim_user , movie_name)
    for movie_name in movies_watched_by_user2:
        if movie_name not in movies_watched_by_user1:
            print('Recommendation for userId', x, movie_name)

    
```
